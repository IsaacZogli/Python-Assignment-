## Question 1.

```python
# 1. Storing items and prices in a dictionary.
shop_items = { 
    "Bread": 25.0,
    "Milk": 18.0,
    "Eggs": 2.5,
    "Sugar": 20.0
}
```

```python
# 2. Using a loop and the input function to ask how many of each item a user wants to buy.
quantities = {}

print("Welcome to Isaac's Shop")
```

```python
for item in shop_items.keys():
    quantity = int(input(f"How many {item} do you want to buy?"))
    quantities[item] = quantity
```

```python
# 3. Total Cost Calculation.
subtotal = 0

for item, price in shop_items.items():
    subtotal += price * quantities[item]
```

```python
# 4. Applying Discount If Subtotal > 100 GHS.
discount = 0
if subtotal > 100:
    discount = 0.10 * subtotal

total = subtotal - discount
```

```python
# 5. Receipt Printing.
print("\n----- RECEIPT -----")
for item, price in shop_items.items():
    item_total = price * quantities[item]
    print(f"{item} (x{quantities[item]}): GHS {item_total:.2f}")
print("---------------")
print(f"Subtotal: GHS {subtotal:.2f}")
print(f"Discount: GHS {discount:.2f}")
print(f"Total: GHS {total:.2f}")
print("----------------")
print("Thank you ðŸ™ for shopping with us ðŸ›’!")
```

## Question 2

### Tasks

```python
# 1. Creating A List Of Dictionaries (each dictionary represents a student)
students = [
    {"name": "Ama", "Maths": 78, "Science": 82, "English": 74},
    {"name": "Kojo", "Maths": 65, "Science": 59, "English": 72},
    {"name": "Esi", "Maths": 90, "Science": 85, "English": 88},
    {"name": "Yaw", "Maths": 70, "Science": 75, "English": 68}
]
```

```python
# 2. Calculating Each Student's Average Score And Add It To Their Dictionary.
for student in students:
    average_score = (student['Maths'] + student['Science'] + student['English']) / 3
    student.update({"Average": average_score})
```

```python
# 3. Printing The Results In A Formatted Way Using f-strings.
print("----- STUDENT SCORES -----")
for student in students:
    print(f"{student['name']} = Maths: {student['Maths']}, Science: {student['Science']}, "
          f" English: {student['English']} -> Average: {student['Average']:.2f}")
```

```python
# 4. Determining The Student With The Highest Average Score.
top_student = students[0]

for student in students:
    if student['Average'] > top_student['Average']:
        top_student = student
```

```python
# 5. Display The Student With The Highest Average.
print("\n---- TOP STUDENT ----")
print(f"{top_student['name']} has the highest average score of {top_student['Average']:.2f}")
```

## Question 3

### Tasks

```python
# 1. 
employees = ['Kwame', 'Akua', 'Yaw', 'Afia']
salaries = [3200, 4000, 2800, 5000]
```

```python
# 2. Combining And Mapping The Two Lists Into Dictionary.
salary_dict = {}

for i in range(len(employees)):
    salary_dict[employees[i]] = salaries[i]
```

# Rule:
+7% if salary < 3500,

otherwise +4%

```python
# 3. Adjusting Salaries Using Dictionary Methods The Rules Given.
for name, old_salary in salary_dict.items():
    if old_salary < 3500:
        new_salary = old_salary * 1.07 # 7% increase
    else:
        new_salary = old_salary * 1.04 # 4% increase

    # Using dictionary method .update() to modify salary
    salary_dict.update({name: new_salary})

    print(f"{name} | Old: {old_salary:.2f} | New: {new_salary:.2f}")
```

#### Writing A Program That Counts How Many Times Each Vowel Appears In A Sentence.

```python
# 1. Getting sentence input from the user.
sentence = input("Enter a sentence: ")
```

```python
# 2. Createing a dictionary that stores vowel counts
vowels = {"a": 0, "e": 0, "i": 0, "o": 0, "u": 0}
```

```python
# 3. Looping through each character in the sentence
for char in sentence.lower():  
    if char in vowels:
        vowels[char] += 1
```

```python
# 4. Printing the results
print("\n--- VOWEL COUNTS ---")
for vowel, count in vowels.items():
    print(f"{vowel.upper()}: {count}")
```
